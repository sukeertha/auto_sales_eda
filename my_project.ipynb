# AUTO SALES DATA ANLAYSIS

COLUMN DESCRIPTION

ORDERNUMBER	 : This column represents the unique identification number assigned to each order.

QUANTITYORDERED	: It indicates the number of items ordered in each order.

PRICEEACH :  This column specifies the price of each item in the order.

ORDERLINENUMBER : It represents the line number of each item within an order.

SALES  : 	This column denotes the total sales amount for each order, which is calculated by multiplying the quantity ordered by the price of each item.

ORDERDATE : It denotes the date on which the order was placed.

DAYS_SINCE_LASTORDER : This column represents the number of days that have passed since the last order for each customer. It can be used to analyze customer purchasing patterns.

STATUS	: It indicates the status of the order, such as "Shipped," "In Process," "Cancelled," "Disputed," "On Hold," or "Resolved."

PRODUCTLINE	: This column specifies the product line categories to which each item belongs.

MSRP :	It stands for Manufacturer's Suggested Retail Price and represents the suggested selling price for each item.

PRODUCTCODE :	This column represents the unique code assigned to each product.

CUSTOMERNAME : 	It denotes the name of the customer who placed the order.

PHONE :	This column contains the contact phone number for the customer.

ADDRESSLINE1 : It represents the first line of the customer's address.

CITY : This column specifies the city where the customer is located.

POSTALCODE : It denotes the postal code or ZIP code associated with the customer's address.

COUNTRY : This column indicates the country where the customer is located.

CONTACTLASTNAME	: It represents the last name of the contact person associated with the customer.

CONTACTFIRSTNAME : This column denotes the first name of the contact person associated with the customer.

DEALSIZE : It indicates the size of the deal or order, which are the categories "Small," "Medium," or "Large."

# DATA CLEANING

```python
# checking null values 

df.isna().sum()
```

```python
# changing the datatype of of column ORDERDATE

df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
# Checking the duplicates in the dataset

df.duplicated().sum()
```

```python
# Replacing the spaces with underscore

df['STATUS'] = df['STATUS'].replace(['On Hold','In Process'],['On_Hold','In_Process'])
```

Checking the outlier in the dataset using box plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(df[['ORDERNUMBER','QUANTITYORDERED','PRICEEACH','ORDERLINENUMBER','SALES','DAYS_SINCE_LASTORDER','MSRP']])
plt.show()
```

Removing the outliers

```python
cols = ['QUANTITYORDERED','PRICEEACH','SALES','MSRP']
outlier_indices = set()
for col in cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 -(1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)
    outlier = df[(df[col] < lower_bound) | (df[col] > upper_bound)].index
    outlier_indices.update(outlier)
df.drop(index=outlier_indices,inplace=True)
```

checking outlier removed

```python
plt.figure(figsize=(12, 6))
sns.boxplot(df[['ORDERNUMBER','QUANTITYORDERED','PRICEEACH','ORDERLINENUMBER','SALES','DAYS_SINCE_LASTORDER','MSRP']])
plt.show()
```

```python
#deleting unneccesary columns
df.drop(['PHONE','ADDRESSLINE1','CONTACTLASTNAME','CONTACTFIRSTNAME'],axis=1,inplace=True)
```

# DATA ANALYSIS

1. Which product line has highest number of orders

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='PRODUCTLINE' ,palette="mako",hue='PRODUCTLINE',legend=False)
plt.xticks(rotation = 45)
plt.xlabel('Product lines')
plt.ylabel('total orders')
plt.title('Productline VS Total order')
plt.show()
```

2. Total order in each status category

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='STATUS' ,palette="mako",hue='STATUS',legend=False)
plt.xticks(rotation = 45)
plt.xlabel('Status')
plt.ylabel('total orders')
plt.title('Total order in each status category')
plt.show()
```

3. Distribution of sales

```python
sns.histplot(df['SALES'],bins=5,kde=True)
plt.xlabel('SALES')
plt.ylabel('count')
plt.title('Distribution of sales')
plt.show()
```

4. Relationship between QUANTITYORDERED and SALES

```python
sns.scatterplot(data=df,x='QUANTITYORDERED',y='SALES')
plt.show()
```

5. TOTAL SALES FOR EACH STATUS CATEGORY

```python
total_sales = df.groupby('STATUS')['SALES'].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x = total_sales.index,y=total_sales.values / 1000 , palette ='mako',hue=total_sales.index,legend=False)
plt.xlabel('Order Status')
plt.ylabel('Total Sales')
plt.title('Total sales for each sales category')
plt.show()
```

6. which product line has highest sales

```python
plt.figure(figsize=(12,6))
sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
sns.barplot(x = sales.index, y= sales.values / 1000, palette="mako",hue=sales.index,legend=False)
plt.xlabel('Product line')
plt.ylabel('Sales')
plt.title('Product line VS Total Sales')
plt.show()
```

7. Which product line has highest MSRP

```python
plt.figure(figsize=(12,6))
sales_msrp = df.groupby('PRODUCTLINE')['MSRP'].sum().sort_values(ascending=False)
sns.barplot(x=sales_msrp.index, y= sales_msrp.values,palette='mako',hue=sales_msrp.index,legend=False)
plt.xlabel('Productline')
plt.ylabel('MSRP')
plt.title('Product line VS MSRP')
plt.show()
```

8. Top 10 Country  has highest total sales 

```python
country_sales = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=country_sales.index, y= country_sales.values /1000 , palette='mako',hue=country_sales.index,legend=False)
plt.xticks(rotation = 45)
plt.xlabel('country')
plt.ylabel('sales')
plt.title('Top 10 countries based on sales')
plt.show()
```

9. TOP 5 CUSTOMERS WHO HAS HIGHEST QUANTITY OF ORDERS

```python
customer = df.groupby('CUSTOMERNAME')['QUANTITYORDERED'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(12,6))
sns.barplot(x=customer.index,y=customer.values,palette='mako',hue=customer.index,legend=False)
plt.xlabel('Customer Name')
plt.ylabel('Quantity of orders')
plt.title('Top 5 customers based on quantity of orders')
plt.xticks(rotation=45)
plt.show()
```

10. Top 10 most frequent customers

```python
df.groupby('CUSTOMERNAME')['ORDERNUMBER'].nunique().sort_values(ascending=False).head(10).reset_index(name='UniqueOrders')
```

11. TOP 5 Customers by sales amount

```python
customer_sales = df.groupby('CUSTOMERNAME')['SALES'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(12,6))
sns.barplot(x=customer_sales.index,y=customer_sales.values,palette='mako',hue=customer_sales.index,legend=False)
plt.xlabel('customer')
plt.ylabel('sales')
plt.title('Top 5 customers by sales amount')
plt.xticks(rotation = 45)
plt.show()
```

12. which deal size has the highest sales

```python
sales_by_deals = df.groupby('DEALSIZE')['SALES'].sum().sort_values(ascending=False)
plt.figure(figsize=(12,6))
sns.barplot(x= sales_by_deals.index,y= sales_by_deals.values / 1000 ,palette='mako',hue=sales_by_deals.index,legend=False)
plt.xlabel('Deal Size')
plt.ylabel('Sales')
plt.title('Deal Size VS Sales')
plt.show()
```

13. Average sales percentage  for each year

```python
year_sales = df.groupby('Year')['SALES'].mean().reset_index()
year_sales
plt.figure(figsize=(6,6))
plt.pie(year_sales['SALES'],labels=year_sales['Year'],autopct='%1.1f%%',colors=['#EC6B56','#FFC154','#47B39C'],shadow=True,explode=[0,0,0.2])
plt.title('Average sales percentage for each year')
plt.show()
```
